```python
import pandas as pd
import matplotlib.pyplot as plt

# 載入 CSV 檔案
file_path = 'raw_output.csv'  # 替換成您的 CSV 檔案路徑
data = pd.read_csv(file_path)
```

```python
plt.figure(figsize=(12, 8))
for column in data.columns:
    plt.plot(data[column], label=column)
plt.fill_between(data.index, data['Truth'], color='yellow', alpha=0.3)
plt.ylim(120, None)  # None 表示自動根據最大值調整
plt.title('All close price predict')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.grid(False)
plt.show()
```

```python
import numpy as np

plot_columns = [4, 5, 6, 9]
x_axis_column = 10  # 第 11 欄為日期

x = data.iloc[:, x_axis_column]
xticks_index = np.linspace(0, len(x) - 1, 5, dtype=int)  # 5 個等距點位

plt.figure(figsize=(12, 8))

# 畫其他預測線
for idx in plot_columns:
    plt.plot(x, data.iloc[:, idx], label=data.columns[idx])

# 填色 + 黃色 Truth 線條
plt.fill_between(x, data['Truth'], color='yellow',alpha=0.4, label='Truth ')

plt.ylim(140, None)
plt.title('Close Price Predictions')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(False)

# 設定 x 軸刻度為五個日期
plt.xticks(x[xticks_index], rotation=30)

plt.show()
```
