```python
import pandas as pd
import matplotlib.pyplot as plt

# 載入 CSV 檔案
file_path = 'output.csv'  # 替換成您的 CSV 檔案路徑
data = pd.read_csv(file_path)
```

```python
plt.figure(figsize=(12, 8))
for column in data.columns:
    plt.plot(data[column], label=column)
plt.fill_between(data.index, data['True price'], color='yellow', alpha=0.3)
plt.ylim(400, None)  # None 表示自動根據最大值調整
plt.title('All close price predict')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.grid(False)
plt.show()
```
