```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("./2025預測結果.csv")
x = range(len(df))
df
```

```python
# 畫圖

# 確保欄位名稱無空格
df.columns = df.columns.str.strip()

# 若日期不是字串，可轉成字串以便顯示
df['日期'] = df['日期'].astype(str)
plt.figure(figsize=(12, 6))
plt.plot(df['日期'], df['真實'], label='Truth', marker='o')
plt.plot(df['日期'], df['預測'], label='Prediction', linestyle='--', marker='x')

# 標示買入點
buy_signals = df[df['買或賣'] == 1]
plt.scatter(buy_signals.index, buy_signals['真實'], color='green', marker='^', s=100, label='buy')

# 標示賣出點
sell_signals = df[df['買或賣'] == -1]
plt.scatter(sell_signals.index, sell_signals['真實'], color='red', marker='v', s=100, label='sell')

plt.xlabel('Date')
plt.ylabel('Value')
plt.title('AAPL recent backtesting')
plt.legend()
plt.grid(False)
plt.tight_layout()
plt.show()
```
